```python
#Finding the relation between time of the day and accidents
```

```python
df.Start_Time
```

```python
df.Start_Time=pd.to_datetime(df.Start_Time)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
 
sns.displot(df.Start_Time.dt.hour, bins=24, stat='density', kde=False, label="Rate of accidents per day in US")
plt.grid() 
plt.show()
```

```python
#Insights:
#The rate of accidents is higher in the mornings between 7AM-9AM
#The rate of accidents is highest in the evening between 3PM-5PM

```

```python
#Finding if the distribution is same for the 7 days of the week:
```

```python

sns.displot(df.Start_Time.dt.dayofweek, bins=7, stat='density', kde=False)
plt.grid()
```

```python
#Hence, the rate of accidents is higher during the weekdays and smaller during weekends.
```

```python
#Finding if the distribution of accidents throughout the day is same during weekdays and weekends?
```

```python
sunday_data=df.Start_Time[df.Start_Time.dt.dayofweek==6]
```

```python
sns.displot(sunday_data.dt.hour, bins=24, stat='density', kde=False)
plt.grid()
```

```python
#Interestingly, The most accidents on Sundays were between 2PM-4PM unlike weekdays.
```

```python
#To compare this data with Monday:
```

```python
monday_data=df.Start_Time[df.Start_Time.dt.dayofweek==0]
```

```python
sns.displot(monday_data.dt.hour, bins=24, stat='density', kde=False)
plt.grid()
```

```python
#Now, we can confirm that the accidents follow one pattern during Weekdays and the other in Weekends.
```
